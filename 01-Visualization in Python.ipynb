# Pandas
Installation:
```bash
conda install pandas
```

http://pandas.pydata.org/

````python
import pandas as pd
data = pd.read_csv('iris.csv')
data.plot.scatter('petal_length', 'petal_width')
````

# Seaborn 
Installation:
```bash
conda install seaborn
```

https://seaborn.pydata.org/

````python
import seaborn
seaborn.pairplot(data, hue='species')
````

# Bokeh
```
conda install bokeh
```

https://bokeh.pydata.org

````python
from bokeh.plotting import figure, show
from bokeh.io import  output_notebook

output_notebook()
p = figure()
p.circle(data.petal_length, data.petal_width)
show(p)
````

# Plotly
```
conda install plotly
```

https://plot.ly/

````python
from plotly.graph_objs import Scatter
from plotly.offline import iplot, init_notebook_mode

init_notebook_mode(connected=True)

p = Scatter(x=data.petal_length, y=data.petal_width, mode='markers')
iplot([p])
````

# Altair
Installation:
```bash
conda install altair --channel conda-forge
```

https://altair-viz.github.io/

````python
import altair as alt

data = alt.load_dataset('cars')

chart = alt.Chart(data).mark_circle().encode(
    x='Horsepower',
    y='Miles_per_Gallon',
    color='Origin',
)

chart
````
